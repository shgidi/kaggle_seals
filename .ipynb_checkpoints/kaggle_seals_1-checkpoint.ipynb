create bounding boxes

```python
def calc_bbs(dotted, raw): 
    diff = cv2.absdiff(dotted, raw)
    gray = cv2.cvtColor(diff, cv2.COLOR_RGB2GRAY)
    ret,th1 = cv2.threshold(gray,0,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)

    cnts = cv2.findContours(th1.copy(), cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)[-2]
    print("Sea Lions Found: {}".format(len(cnts)))

    x, y = [], []

    lion_patches = []

    radius=25
    for loc in cnts:
        ((xx, yy), _) = cv2.minEnclosingCircle(loc)

        # store patches of some sea lions
        if xx > 10 and xx < gray.shape[1] - 10:
            lion_patches.append(cropped_raw[yy-radius:yy+radius, xx-radius:xx+radius])

        x.append(xx)
        y.append(yy)

    x = np.array(x)
    y = np.array(y)
    return [x,y]
```

```python
%cd /mnt/kaggle_seals//
from keras.applications.imagenet_utils import preprocess_input
anot_id_list=[]
anot_content_list=[]
train_inputs=[]
train_targets=[]
for file,file_dot in zip(train_imgs,train_dot_imgs):
    img=np.array(Image.open(file))
    img_dot=np.array(Image.open(file_dot))
    size=img.shape[:2]
    x,y=calc_bbs(img,img_dot) # x and y are arrays -f 4 X seals

    for x_i,y_i in zip(x,y):
        cropped_img=img[(x_i-150):(x_i+150),y_i-150:y_i+150,:]
        
        if cropped_img.shape[0]==300 and cropped_img.shape[1]==300:
            anot_id=file
            anot_data=[[(x_i-15)/size[0],(y_i-15)/size[1],(x_i+15)/size[0],(y_i+15)/size[1],1]]
            anot_id_list.append(anot_id)
            anot_content_list.append(anot_data)

            z = anot_data.copy()
            #if train and self.do_crop:
            #    img, y = self.random_sized_crop(img, y)
            #img = imresize(img, [300,300]).astype('float32')

            z_processed = bbox_util.assign_boxes(np.array(z))
            train_inputs.append(cropped_img)             
            train_targets.append(z_processed)
        
train_target_array=np.array(train_targets)
train_array=np.array(train_inputs)
X_train=preprocess_input(train_array)
```
